```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales": [120,150,170,140,180,200],
    "Profit": [20,30,35,25,40,50],
    "Expenses": [100,120,135,115,140,150]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.figure()
plt.plot(df["Month"], df["Sales"])
plt.title("Sales Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure()
plt.bar(df["Month"], df["Profit"])
plt.title("Profit Comparison")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure()
plt.pie(df["Sales"], labels=df["Month"], autopct="%1.1f%%")
plt.title("Sales Share")
plt.show()
```

```python
plt.figure()
plt.hist(df["Sales"], bins=5)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit Relationship")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure()
plt.boxplot(df["Profit"])
plt.title("Profit Spread")
plt.show()
```
