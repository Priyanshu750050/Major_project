```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# dataset load
rt=pd.read_csv(r"stolen_vehicles.csv")
rt.head()
```

```python
# plot scatter
plt.scatter(rt["date_stolen"],rt["location_id"])
plt.xlabel("date_stolen")
plt.ylabel("location_id")
plt.title("scatter plot")
plt.show()
```

```python
# bar plot
rt["color"].value_counts().plot(kind="bar")
plt.xlabel("color")
plt.ylabel("count")
plt.title("bar plot")
plt.show()
```

```python
# box plot
rt.boxplot(column="location_id",by="color")
plt.xlabel("color")
plt.ylabel("location_id")
plt.title("box plot")
plt.show()
```

```python
# pie chart
rt["color"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("pie chart")
plt.show()
```

```python
# histogram
rt["location_id"].plot(kind="hist",bins=10)
plt.xlabel("location_id")
plt.ylabel("frequency")
plt.title("histogram")
plt.show()
```

```python
# line plot
rt.groupby("color")["location_id"].mean().plot(kind="line")
plt.xlabel("color")
plt.ylabel("mean location_id")
plt.title("line plot")
plt.show()
```

```python
# data visualization
rt.hist(column="location_id",by="color",figsize=(10,5))
```

```python
# distribution plot
sns.displot(rt["location_id"],kde=True)
```
