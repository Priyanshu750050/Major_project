```python
df["is_Apple"] = df["Battery_mAh"]>4000
if df["is_Apple"].value_counts()[True]>df["is_Apple"].value_counts()[False]:
    print("Most of the phones in the dataset have a battery capacity greater than 4000 mAh.")
else:
    print("Most of the phones in the dataset have a battery capacity less than or equal to 4000 mAh.")

print(df["is_Apple"].value_counts())
```

```python
df["Price_INR"] = df["Price_USD"]*93.28
df
```

```python
# Distribution of Smartphone Prices in INR.
budget = 250000 * 0.10
affordable_phones = df[df["Price_INR"] <= budget]
print(affordable_phones[["Brand", "Model", "Price_INR"]])
plt.figure(figsize=(10,6))
sns.histplot(df["Price_INR"], bins=20, kde=True)
plt.title("Distribution of Smartphone Prices in INR")
plt.xlabel("Price (INR)")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Battery Capacity vs Price.
sns.scatterplot(data=df, x="Battery_mAh", y="Price_INR")
plt.title("Battery Capacity vs Price")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Price (INR)")
plt.show()
```

```python
# Price Distribution by Brand.
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Brand", y="Price_INR")
plt.title("Price Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Price (INR)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Battery_mAh", "Price_INR", "RAM_GB check correlation.
plt.figure(figsize=(10,6))
sns.pairplot(df[["Battery_mAh", "Price_INR", "RAM_GB"]])
plt.show()
```

```python
# Correlation Heatmap.
sns.heatmap(df[["Battery_mAh", "Price_INR", "RAM_GB"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Market Share of Smartphone Brands.
plt.figure(figsize=(10,6))
plt.pie(df["Brand"].value_counts(), labels=df["Brand"].value_counts().index, autopct="%1.1f%%")
plt.title("Market Share of Smartphone Brands")
plt.axis("equal")
plt.show()
```

```python
# count of phones by brand most salling company.
most_sales_brand = df["Brand"].value_counts().idxmax()
print(f"The brand with the most sales is: {most_sales_brand}")
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Brand", order=df["Brand"].value_counts().index[:])
plt.title("Count of Phones by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
