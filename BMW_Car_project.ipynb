```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\mrash\Downloads\archive (3)\bmw.csv")
print(df.head())
```

```python
df.isnull().sum()
df.info()
df.describe()
```

```python
df["fuelType"].value_counts()
df[df["fuelType"] == "Electric"]["model"]
df[df["year"]==2015]["model"]
```

```python
df["multiplication"]=df["mileage"]*df["tax"]
print(df["multiplication"])   
```

```python
df["fuelType"].value_counts()
plt.pie(df["fuelType"].value_counts(), autopct='%1.1f%%')
plt.legend(df["fuelType"].value_counts().index)
plt.title("Fuel Type Distribution") 
plt.show()
```

```python
plt.bar(df["transmission"].value_counts().index, df["transmission"].value_counts().values)
plt.xlabel("Transmission")
plt.ylabel("Count")
plt.title("Transmission Type Distribution")
plt.show()
```

```python
# bar chat to campare the count of different car models with year
plt.figure(figsize=(12,6))
sns.countplot(x="model", hue="year", data=df)
plt.xticks(rotation=90)
plt.title("Count of Car Models by Year")
plt.show()
```

```python
# Scatter plot of price vs mileage
plt.figure(figsize=(10,6))
sns.scatterplot(x="mileage", y="price", data=df)
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.title("Price vs Mileage")
plt.show()
```

```python
#  histplot of price vs fuelType
plt.figure(figsize=(10,6))
sns.histplot(x="fuelType", y="price", data=df)
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.title("Price vs Fuel Type")
plt.show()
```

```python
# multiple ghrap to compare the price distribution of different car models
plt.figure(figsize=(12,6))
sns.boxplot(x="model", y="price", data=df)
plt.xticks(rotation=90)
plt.title("Price Distribution by Car Model")
plt.show()
```

```python
# hitmap to show the correlation between different features
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# use multiple sns chart to show the relationship between different features
sns.pairplot(df.select_dtypes(include=[np.number])) 
plt.show()
```
